```python
import artm
import os
import matplotlib.pyplot as plt
```

```python
path = 'C:\\NIVC\\Nivc_BigARTM_corpus\\unary_comm\\'
subd = "golosislamacom"
batch_vectorizer = artm.BatchVectorizer(data_path=path + "\\" + subd + "\\" + "batches_pos",
                                        data_format='batches')
```

```python
lda = artm.LDA(num_topics=200, alpha=0.02, beta=0.02,
               num_document_passes=1, dictionary=batch_vectorizer.dictionary,
               cache_theta=True)
```

```python
lda.fit_offline(batch_vectorizer=batch_vectorizer, num_collection_passes=5)
```

```python
print lda.sparsity_phi_last_value
print lda.sparsity_theta_last_value
print lda.perplexity_value
```

```python
top_tokens = lda.get_top_tokens(num_tokens=100)
for i, token_list in enumerate(top_tokens):
    print 'Topic #{0}: {1}'.format(i, token_list)
```
