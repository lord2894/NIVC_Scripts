```python
import logging, gensim
BigARTMCorpusDir = 'C:\\NIVC\\Nivc_GenSim_corpus\\binary_comm\\'
subd = "islamreviewru"
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# load id->word mapping (the dictionary), one of the results of step 2 above
id2word = gensim.corpora.Dictionary.load(BigARTMCorpusDir+subd+"\\mm_pos\\"+subd+'.dict')
# load corpus iterator
mm = gensim.corpora.MmCorpus(BigARTMCorpusDir+subd+"\\mm_pos\\"+subd+'.mm')

print(mm)
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=100, update_every=1, chunksize=100, passes=1)
```

```python
import os
if not os.path.exists(BigARTMCorpusDir+subd+"\\LDA_model"):
    os.makedirs(BigARTMCorpusDir+subd+"\\LDA_model", mode=0o777)
if not os.path.exists(BigARTMCorpusDir+subd+"\\LDA_topics"):
    os.makedirs(BigARTMCorpusDir+subd+"\\LDA_topics", mode=0o777)
```

```python
lda.save(BigARTMCorpusDir+subd+"\\LDA_model\\"+subd+".model")
```

```python
topics100 = lda.show_topics(100,15)
topics20 = lda.show_topics(20,15)
```

```python
import io
with io.open(BigARTMCorpusDir+subd+"\\LDA_topics"+'\\20.topics','w',encoding='utf8') as outputTopics:
    try:
        j = 0
        outputTopics.write(u'{\n')
        for topic in topics20:
            outputTopics.write(u'\"'+unicode(topic[0])+u'\" : {')
            words = topic[1].split("+")
            i = 0
            for word in words:
                wordNum = word.split("*")
                if i != len(words)-1:
                    outputTopics.write(u'\"'+wordNum[1]+u'\" : '+wordNum[0]+u',')
                else:
                    outputTopics.write(u'\"'+wordNum[1]+u'\" : '+wordNum[0]+u'}')
                i+=1
            if j != len(topics20)-1:
                outputTopics.write(u',\n')
            else:
                outputTopics.write(u'}')
            j+=1
    except Exception as e:
        print(e)
    outputTopics.close()
```
